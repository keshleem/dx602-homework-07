**Concept:** *List data type, list comprehension.*

**Task:**

Write a function `stdev(values)` that takes a 1D Python list of numbers and returns the population standard deviation.

**Sample Output:**


```
stdev([1, 2, 3]) should return approximately 0.8164965809

stdev([5, 5, 5, 5]) should return 0.0
```


````python
def stdev(values):
  '''
  takes 1D Python list of numbers and returns the population standard deviation
  '''
  # gets the mean of the list
  mean = sum(values) / len(values)
  # get the variance
  variance = sum((x - mean) ** 2 for x in values) / len(values)
  # returns the variance
  return variance ** 0.5
````

Write a function `sum_list(a, b)` that takes two same-length 1D lists and returns a new list containing the element-wise sums.

If the lengths differ, print a message indicating incompatible lengths in this format: Incompatible lengths: {len(a)} vs {len(b)}.
Use list comprehension for this, not numpy.

**Sample Output:**

```
sum_list([2, 4, 6], [1, 3, 5]) → [3, 7, 11]
```


````python
def sum_list(a, b):
  '''
  takes two same-length 1D lists and returns a new list containing the element-wise sums
  '''
  # if the length of list a is not equal to list b
  if len(a) != len(b):
    # print the following 
    print(f'Incompatible lengths: {len(a)} vs {len(b)}')
  # if the length is equal to b
  else:
    # return the sum of the index of both lists for the range of the length of a
    return [a[i] + b[i] for i in range(len(a))]
````

**Problem 6**

**Concept:** *Array operations, numpy functions and methods.*

**Task:**

Write a function `softmax(z)` that takes in a 1D numpy array of values $z_i$, and returns an array composed of $\frac{e^{z_i}}{\sum e^{z_i}}$.

The output array shape should be the same as the input array shape.

This formula is the [softmax](https://en.wikipedia.org/wiki/Softmax_function) formula used to adjust raw machine learning outputs into more suitable probability estimates.

**Example Function Calls:**

`softmax(np.array([0,0]))` returns `np.array([0.5, 0.5])`

`softmax(np.array([-1, 0, 1]))` returns `np.array([0.09003057 0.24472847 0.66524096])`


````python
# TO DO: WRITE YOUR SOLUTION HERE
def softmax(z):
  ''' 
  takes in a 1D numpy array of values z and returns an array'''
  # e^x to each of the values in the array
  z = np.exp(z)
  # divides each of the exponential values by sum to get the softmax prob, the values equal to 1
  return z / np.sum(z)
````

**Hint:**

**Example Function Calls:**

`np.array([[1. 4. 7.], [2. 5. 8.], [3. 6. 9.]])`



````python
# TO DO: WRITE YOUR SOLUTION HERE
def three_stats(x):
  '''
  takes an n x k array and returns a 3 by N array of min, meds, max'''
  # find min value in each row
  minimum_row = np.min(x, axis=1)
   # find median value in each row
  median_row = np.median(x, axis=1)
   # find max value in each row
  max_row = np.max(x, axis=1)
  # returns the min, med, and max rows
  return np.vstack([minimum_row, median_row, max_row])
````

**Problem 12**

**Concept:** *File reading, parsing data values, array creation.*

**Task:**
Write the function `read_csv_to_array(filename)` that reads a CSV file, discards the header row, converts each element to a floating-point number, and returns a 2D numpy array.

**Sample Output:**

If file contains:

```
foo,bar
2,4
6,8
```

The function should return:
```
np.array([[2. 4.][6. 8.]])
```

````python
# TO DO: WRITE YOUR SOLUTION HERE
def read_csv_to_array(filename):
  ''' 
  reads CSV, discards head row, converts each elemnt to floating-point number and returns 2D numpy array
  '''
  with open(filename, 'r') as file:
    lines = file.readlines()

    data = []
    for line in lines[1:]:
      line = line.strip()
      if line:
        row = [float(x) for x in line.split(',')]
        data.append(row)

    return np.array(data)
````

**Problem 13**

**Concept:** *File reading, parsing data values, array creation.*


**Task:**

Write the function `get_column_stats(filename)` that a CSV file, discards the header row, and returns a 2D numpy array where the rows contain in order: the means, standard deviations, minimums, and maximums, of each column.

**Sample Output:**

If file contains:
```
a,b,c
1,2,3
4,5,6
```
The resulting columns means are

```np.array([[2.5, 3.5, 4.5], [1.5, 1.5, 1.5], [1, 2, 3], [4, 5, 6]])```


````python
# TO DO: WRITE YOUR SOLUTION HERE
def get_column_stats(filename):
  ''' 
  discards header row, returns 2D numpy array where the rows contain in order - means, standard deviations, min, max of each column
  '''
  data = read_csv_to_array(filename)
  means = np.mean(data, axis=0)
  std_devs = np.std(data, axis=0)
  minimums = np.min(data, axis=0)
  maximums = np.max(data, axis=0)
  return np.vstack([means, std_devs, minimums, maximums])
````
